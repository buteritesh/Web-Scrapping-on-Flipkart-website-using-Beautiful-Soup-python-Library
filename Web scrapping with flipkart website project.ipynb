# 1)get html

```python
r=requests.get(url)
htmlcontent=r.content
htmlcontent
```

# Now we going to prasing or scrapping this html data with the module beautifulsoup

```python
soup=BeautifulSoup(htmlcontent,'html.parser')
print(soup.prettify)
```

# Now we find the html title

```python
title=soup.title
title
```
